Tierra (149.597.870 km) y Júpiter (778.547.200 km).

```python
# Crear variables para almacenar las dos distancias
# ¡Asegúrate de quitar las comas!

tierra = 149597870
jupiter = 778547200
```

Ejercicio 2: convierte cadenas en números y usa valores absolutos

```python
# Almacenar las entradas del usuario
#Pista: variable = input("¿Cuál es tu nombre?")
planeta_uno = input("Indica la distancia del primer planeta")
planeta_dos = input("Indica la distancia del segundo planeta")
```

```python
# Convierte las cadenas de ambos planetas a números enteros
planeta_uno = int(planeta_uno)
planeta_dos = int(planeta_dos)
```

```python
# Realizar el cálculo y determinar el valor absoluto
calculo_distancia = abs(planeta_uno - planeta_dos)
print(f'Distancia en KM: {calculo_distancia}')
# Convertir de KM a Millas
print(f'Distancia en Millas: {calculo_distancia * 0.621}')
```
