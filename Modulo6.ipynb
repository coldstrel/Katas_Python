Ejercicio 2: Trabajando con datos de una lista

```python
# Lista de planetas
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Neptune']
```

```python
# Solicitamos el nombre de un planeta *Pista:  input()*
nombre_planeta = input("Ingresa el nombre de un planeta, inicia el nombre con mayusculas")
```

```python
# Busca el planeta en la lista
print(planets.index(nombre_planeta))
print(planets[planets.index(nombre_planeta)])
```

```python
# Muestra los planetas más cercanos al sol
print(f'Planetas mas cercanos al sol que {nombre_planeta}')
for planet in range(planets.index(nombre_planeta)):
    print(planets[planet])
```

```python
# Muestra los planetas más lejanos al sol
print(f'Planetas mas lejanos al sol que {nombre_planeta}')
print(planets[planets.index(nombre_planeta) + 1 : -1])
```
