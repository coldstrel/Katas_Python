Ejercicio 1: Creacion de diccionarios de Python

```python
# Crea un diccionario llamado planet con los datos propuestos
planet = {
    'name' : 'Mars',
    'moons' : '2'
}
```

```python
# Muestra el nombre del planeta y el número de lunas que tiene.
print(f'Planeta: {planet.get("name")} \nLunas: {planet.get("moons")}' )
```

```python
# Agrega la clave circunferencia con los datos proporcionados previamente
planet['circunferencia'] = {
    'polar' : 6752,
    'equatorial' : 6792
}
```

```python
print(f'Planeta: {planet.get("name")} \nLunas: {planet.get("moons")} \nCircunferencia: \n\tPolar = {planet["circunferencia"]["polar"]} \n\tEcuatorial={planet["circunferencia"]["equatorial"]}')
```

Ejercicio 2: Programación dinámica con diccionarios

```python
# Planets and moons

planet_moons = {
    'mercury': 0,
    'venus': 0,
    'earth': 1,
    'mars': 2,
    'jupiter': 79,
    'saturn': 82,
    'uranus': 27,
    'neptune': 14,
    'pluto': 5,
    'haumea': 2,
    'makemake': 1,
    'eris': 1
}
```

```python
# Añade el código para determinar el número de lunas.
moons = planet_moons.values()
planets = len(planet_moons)
```

```python
# Agrega el código para contar el número de lunas. 
total_moon = 0
for moon in moons:
    total_moon = total_moon + moon

print(f'Lunas totales: {total_moon}')

# Finalmente calcule el promedio dividiendo total_moons por planets e imprimiendo los resultados.
print(f'Promedio de lunas por planeta {round(total_moon / planets)}')

```
