Ejercicio 1: Trabajar con argumentos en funciones

```python
# Función para leer 3 tanques de combustible y muestre el promedio
def tanques(tanque1, tanque2, tanque3):
    return (f'La capacidad del tanque 1 es: {tanque1} litros \nLa capacidad del tanque 2 es: {tanque2} litros \nLa capacidad del tanque 3 es: {tanque3} litros')
```

```python
# Llamamos a la función que genera el reporte print(funcion(tanque1, tanque2, tanque3))
tanque1 = int(input("ingresa la capacidad del primer tanque: "))
tanque2 = int(input("ingresa la capacidad del segundo tanque: "))
tanque3 = int(input("ingresa la capacidad del tercer tanque: "))
print(tanques(tanque1,tanque2,tanque3))
```

```python
# Función promedio 
def promedio(tanque1, tanque2, tanque3):
    average = (tanque1 + tanque2 + tanque3) / 3
    return (f'El promedio de la capacidad de los tanques es: {average}')
```

```python
promedio(tanque1,tanque2,tanque3)
```

```python
# Actualiza la función
def tanques(tanque1, tanque2, tanque3):
    print(promedio(tanque1, tanque2, tanque3))
    print(f'La capacidad del tanque 1 es: {tanque1} litros \nLa capacidad del tanque 2 es: {tanque2} litros \nLa capacidad del tanque 3 es: {tanque3} litros')
```

```python
# Escribe tu nueva función de reporte considerando lo anterior
def informe_preciso(destino, *time_travel, **tanque):
    print(f"""
    Tenemos como destino: {destino}
    Nos tomara {sum(time_travel)} minutos llegar
    Tenemos {sum(tanque.values())} L para llegar""")
```

```python
# Debido a que el combustible que queda en los tanques es específico de cada tanque, actualiza la función para usar el nombre de cada tanque en el informe:
# Escribe tu nueva función
def informe_preciso(destino, *time_travel, **tanque):
    print(f"""
    Tenemos como destino: {destino}
    Nos tomara {sum(time_travel)} minutos llegar
    """)

    for nombre_tanque, capacidad in tanque.items():
        print(f"""
        El tanque {nombre_tanque} tienen una capacidad de: {capacidad}L""")
```

```python
informe_preciso('tierra', 900,40,22, tanque1=200, tanque2=321, tanque3=333)
```
